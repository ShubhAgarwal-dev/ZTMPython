Exercise of python --> Find duplicates in list

```python
some_list = ['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n']
# to sort this list
def sort_list(somelist : list):
    duplicates = []
    for i in somelist:
        if somelist.count(i)>1:
            if i not in duplicates:
                duplicates.append(i)
    
    return duplicates

print(sort_list(some_list))
```

Exercise of Python --> Classes

```python
class Pets():
    animals = []
    
    def __init__(self, animals: list):
        self._animals = animals

    def walk(self):
        for animal in self._animals:
            print(animal.walk())


class Cat():
    is_lazy = True

    def __init__(self, name: str, age: int = 3):
        self._name = name
        self._age = age

    def walk(self):
        return f"{self._name} is walking around"


class Simon(Cat):

    def sing(self, sound):
        return f"{sound}"

# I am in edit mode
class Sally(Cat):

    def sing(self, sound):
        return f"{sound}"


class Samuel(Cat):

    def sing(self, sound):
        return f"{sound}"


simon = Simon("Simon")
sally = Sally("Sally")
samuel = Samuel("Samuel")

my_cats = [simon, sally, samuel]
my_pets = Pets(my_cats)
my_pets.walk()
```

Exercise of python --> Extended List

```python
class SuperList(list):

    def __len__(self) -> int:
        return 1000

my_super_list = SuperList()
print(len(my_super_list))
my_super_list.append(5)
print(my_super_list)
print(isinstance(my_super_list, list)) #To check the instance
print(issubclass(SuperList, list)) # To check the class (blueprint)
```

Find MRO

```python
class X:
    pass

class Y:
    pass

class Z:
    pass

class A(X, Y):
    pass

class B(Y, Z):
    pass

class M(A, B, Z):
    pass

print(M.mro())
# M -> A -> X -> B -> Y -> Z -> Object
```

Exercise of python -> pure function

```python
my_pets = ['sisi', 'bibi', 'titi', 'carla']
#? Capitalize all of pet name and print the list 

my_string = ['a', 'b', 'c', 'd', 'e']
my_number = [5, 4, 3, 2 , 1]
#? Zip list into tupple bur sort list from highest to lowest

score = [73, 20, 65, 19, 76, 100, 88]
#? Filter score that pass over 50%

def capitalize(item: str):
    str.capitalize(item)
    return item

def sorting(li: list):
    li2 = li.copy()
    list.sort(li2)
    return li2

def check_score(item: int):
    return item > 50

print(list(map(capitalize, my_pets)))
print(list(zip(my_string, sorting(my_number))))
print(my_number)
print(list(filter(check_score, score)))
```

Exercise of python -> Decorator

```python
# Create an @authenticated decorator that only allows the function to run is user1 has 'valid' set to True:
user1 = {
    'name': 'Sorna',
    'valid': False #changing this will either run or not run the message_friends function.
}

def authenticated(fn):
    def wrapper(*args, **kwargs):
        if user1["valid"] == True:
            return fn(*args, **kwargs)
        else :
            return "Invalid user"

    return wrapper


@authenticated
def message_friends(user):
    print('message has been sent')

message_friends(user1)
```
