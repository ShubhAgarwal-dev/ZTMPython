Exercise of python --> Find duplicates in list

```python
some_list = ['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n']
# to sort this list
def sort_list(somelist : list):
    duplicates = []
    for i in somelist:
        if somelist.count(i)>1:
            if i not in duplicates:
                duplicates.append(i)
    
    return duplicates

print(sort_list(some_list))
```

Exercise of Python --> Classes

```python
class Pets():
    animals = []
    
    def __init__(self, animals: list):
        self._animals = animals

    def walk(self):
        for animal in self._animals:
            print(animal.walk())


class Cat():
    is_lazy = True

    def __init__(self, name: str, age: int = 3):
        self._name = name
        self._age = age

    def walk(self):
        return f"{self._name} is walking around"


class Simon(Cat):

    def sing(self, sound):
        return f"{sound}"

# I am in edit mode
class Sally(Cat):

    def sing(self, sound):
        return f"{sound}"


class Samuel(Cat):

    def sing(self, sound):
        return f"{sound}"


simon = Simon("Simon")
sally = Sally("Sally")
samuel = Samuel("Samuel")

my_cats = [simon, sally, samuel]
my_pets = Pets(my_cats)
my_pets.walk()
```

Exercise of python -> Decorator

```python
# Create an @authenticated decorator that only allows the function to run is user1 has 'valid' set to True:
user1 = {
    'name': 'Sorna',
    'valid': False #changing this will either run or not run the message_friends function.
}

def authenticated(fn):
    def wrapper(*args, **kwargs):
        if user1["valid"] == True:
            return fn(*args, **kwargs)
        else :
            return "Invalid user"

    return wrapper


@authenticated
def message_friends(user):
    print('message has been sent')

message_friends(user1)
```

Exercise of python -> Hacker News

```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint


def check_by_page(last_page):
    links = []
    subtext = []
    for i in range (1, last_page+1):
        pages = {'p': i}
        responce = requests.get(r'https://news.ycombinator.com/news', params=pages)
        soup = BeautifulSoup(responce.text, 'html.parser')
        links += soup.select('.titlelink')
        subtext += soup.select('.subtext')
    return {
        'links': links,
        'subtext': subtext
    }



def sort_by_votes(hnlist):
    return sorted(hnlist, key=lambda k: k['votes'], reverse=True)


def create_custom_link(book:dict):
    hn = []
    for index, item in enumerate(book['links']):
        title = item.get_text()
        href = item.get('href', None)
        vote = book['subtext'][index].select('.score')
        if len(vote):
            point = int(vote[0].get_text().replace('points', ''))
        else:
            point = 0
        if point>=99:
            hn.append({
                'title': title,
                'link': href,
                'votes': point
            })
    return sort_by_votes(hn)
```

```python
pprint(create_custom_link(check_by_page(10)))
```
